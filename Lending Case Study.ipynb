## Drop all columns that more than 10% null values

```python
drop_cols=df.columns[100*df.isnull().mean() >=10].tolist()
drop_cols
```

```python
df.drop(columns=drop_cols, axis =1, inplace =True)
```

```python
df.info()
```

## Imputing data for below numerical columns with the medians since very few records are missing
- revol_util
- pub_rec_bankruptcies
- tax_liens

```python
df['revol_util'] = df['revol_util'].apply(lambda x : float(str(x).rstrip('%')))
df['int_rate'] = df['int_rate'].apply(lambda x : float(str(x).rstrip('%')))
rvu_mid=df['revol_util'].median()    
df.revol_util.fillna(rvu_mid, inplace=True)
 

```

## Remove rows with null entry for the below categorical columns since the very few records are missing
- last_credit_pull_d
- last_pymnt_d
- emp_title
- emp_length

```python
drop_index=df[(df['last_credit_pull_d'].isnull())|(df['last_pymnt_d'].isnull())
             |(df['emp_title'].isnull())|(df['emp_length'].isnull())].index
```

```python
df.drop(drop_index,inplace=True)
```

## Dropping  columns not used in analysis
- Columns chargeoff_within_12_mths,collections_12_mths_ex_med policy_code,application_type,acc_now_delinq         ,delinq_amnt,tax_liens are dropped because it has only 1 fixed value
- Columns title, zip_code, id, member_id , url dont provide any valueable info
  

```python
df.drop(columns=['inq_last_6mths','out_prncp','collection_recovery_fee',
                 'recoveries','pub_rec_bankruptcies','total_pymnt_inv','total_rec_prncp','total_rec_int',
                 'total_rec_late_fee','out_prncp_inv','total_pymnt',
                 'chargeoff_within_12_mths','pymnt_plan','collections_12_mths_ex_med',
                 'title','zip_code', 'initial_list_status','id', 'url','member_id','policy_code',
                 'application_type','acc_now_delinq','delinq_amnt','tax_liens'],inplace=True)
```

```python
date_field_to_extract_month=['issue_d','earliest_cr_line','last_pymnt_d','last_credit_pull_d']
```

# Converting Date Objects
- breaking down the date object to month and year for futher anlysis

```python
for col in date_field_to_extract_month:
    df[col]= pd.to_datetime(df[col], format='%b-%y')
    df[col+'_mnth']=df[col].dt.month
    df[col+'_year']=df[col].dt.year
```

# Univariate Analysis on categorical columns 
- the data wa split into two parts one for loan_status fully Paid and other with status charged off to better understand the trends between the categories

```python
df_defaulter = df[df.loan_status=='Charged Off']
df_fully_paid = df[df.loan_status=='Fully Paid']
```

```python
for col in cat_cols:
    plt.figure(figsize=(15,8))
    plt.subplot(121)
    uni_vals_def = df_defaulter[col].unique()
    uni_vals_fp = df_fully_paid[col].unique()

    
    sns.countplot(x='loan_status',hue=col,data=df_defaulter.sort_values(by=col),palette="Set2")
    plt.title( col + " distribution in defaulters")
    
    plt.subplot(122)
    sns.countplot(x='loan_status',hue =col,data =df_fully_paid.sort_values(by=col),palette ="Set2")
    plt.title( col + " distribution in non defaulters")    
    
    plt.show()
```
